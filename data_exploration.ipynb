# Project assignment 1: Data Exploration

## Get the data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('housing.csv')
df.head()
```

## Analyze the data quality

```python
print("=== Dataset Info ===")
df.info()

print("\n === Unique Values per Column ===")
print(df.nunique())
```

## Description of each important column

```python
important_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
df_important = df[important_cols]
df_important.info()
```

## Statistical summary of quantitative variables

```python
stats_summary = pd.DataFrame(index=important_cols)

stats_summary['Mean'] = df_important.mean().round(2)
stats_summary['Median'] = df_important.median()
stats_summary['Mode'] = df_important.mode().iloc[0]
stats_summary['SD'] = df_important.std().round(2)
stats_summary['Range'] = df_important.apply(lambda x: f"{x.max()} - {x.min()} (Diff: {x.max() - x.min()})")

print(stats_summary)
```

```python
missing_data = df_important.isnull().sum()
print(missing_data)
```
